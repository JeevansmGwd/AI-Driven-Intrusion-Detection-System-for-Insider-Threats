### FUNCTION TO REMOVE THE SPECIFIED PATTERN IN THE STRING COLUMN

````python

def remove_string_from(dataframe, columnname=None, pattern=None, convert_to_16bit=False):
    # Check if columnname is provided
    if columnname is None:
        print("MISSINGPARAM: Specify columnname='COLUMN' as parameter")
        return dataframe
    
    # Check if pattern is provided
    if pattern is None:
        print("MISSINGPARAM: Specify pattern='THING-TO-REMOVE' as parameter")
        return dataframe
    
    # Check if the columnname exists in the dataframe
    if columnname not in dataframe.columns:
        print(f"INVALIDCOLUMN: The column '{columnname}' does not exist in the dataframe")
        return dataframe

    # Check if the column contains string values before applying str.replace
    if not pd.api.types.is_string_dtype(dataframe[columnname]):
        print(f"INVALIDTYPE: The column '{columnname}' is not of string type")
        return dataframe    

    print(f"just removing '{pattern}' from '{columnname}'")
    dataframe[columnname] = dataframe[columnname].str.replace(pattern, '', regex=False)
    
    # Optionally convert to 16-bit integers if required
    if convert_to_16bit:
        # Convert the column to numeric, coercing errors to NaN, and then to 16-bit integers
        dataframe[columnname] = pd.to_numeric(dataframe[columnname], errors='coerce').astype('Int16')

    return dataframe
````
